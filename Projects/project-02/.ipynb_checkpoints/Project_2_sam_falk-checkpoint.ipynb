```python
import pandas as pd
top_billboard=pd.read_csv('/Users/samanthafalk/GA-DSI/Projects/project-02/assets/billboard.csv')
import plotly.plotly as py
import plotly.graph_objs as go
import scipy.stats as stats
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
% matplotlib inline 

from matplotlib.widgets import Cursor

## load in billboard data

## Problem that it only tracks songs that have peaked, not all music that has made it to the billboard- 
##      so it is difficult to make extensive assumptions around the data

## Assume that Rock n Roll is the same as Rock 
## Assume that all artists are properly labeled with the proper genre
## Null Hypothesis: Regardless of length of song, a song will apply to the normal distribution of weeks on Billboard
## Alternate Hypothesis: The shorter the song the longer it lasts on Billboard
```

```python
## Remove empty columns 
top_billboard=top_billboard.iloc[0:,0:72]
```

```python
#change * elements to NaN 
import numpy as np
def nana(x):
    if x=='*':
        x=np.nan
        return x
    else:
        return x


top_billboard=top_billboard.applymap(nana)
```

```python
#clean data for artists
def matchbox(x):
    if x=='matchbox twenty':
        x='Matchbox Twenty'
        return x
    else:
        return x


top_billboard=top_billboard.applymap(matchbox)
```

```python
#clean data for Genre
def RockRoll(x):
    if x=="Rock'n'roll":
        x='Rock'
        return x
    else:
        return x


top_billboard= top_billboard.applymap(RockRoll)
```

```python
#clean data for Genre
def RB(x):
    if x=="R & B":
        x='R&B'
        return x
    else:
        return x


top_billboard= top_billboard.applymap(RB)
```

```python
week_eval=week_eval.T
```

```python
#repeat of hypothesis: longer songs do not last as long on the billboard as shorter songs
#count the amount of weeks each song is on the chart and add it as a column

   
top_billboard['weekscount']=[int(week_eval[x].count()) for x in week_eval]    
```

```python
#check billboard for new column
top_billboard
```

```python
#create new data frame with information to test, sorted by length of song
weeks_songlength= pd.concat([top_billboard['time'], top_billboard['weekscount']], axis=1).reset_index(drop=True)
weeks_songlengthsorted= weeks_songlength.sort('time').reset_index(drop=True)

#find median of song lengths
median= weeks_songlengthsorted['time'][(len(weeks_songlengthsorted['time']))/2]
median

#create new row that denotes if a song is long or short
weeks_songlengthsorted['L_o_S']=["Long" if x > median else "Short" for x in weeks_songlengthsorted['time']]
weeks_songlengthsorted
```

```python
## seperate the info into two data sets (long and short) then find the mean

short_list=weeks_songlengthsorted.loc[weeks_songlengthsorted['L_o_S']=='Short', :]
np.mean(short_list['weekscount'])
```

```python
long_list=weeks_songlengthsorted.loc[weeks_songlengthsorted['L_o_S']=='Long', :]
np.mean(long_list['weekscount'])
```

```python
#then find the difference of the means, this will help validate the ttest
abs(np.mean(long_list['weekscount'])-np.mean(short_list['weekscount']))
```

```python
#ttest of long and short

stats.ttest_ind(short_list['weekscount'], long_list['weekscount'])
```

P value is large, so we fail to reject the null. AKA there is no statistical significance between these two data sets

```python
##Shuffle TEST, run through the data 10K times, by randomly shuffling the data each time and then 
###finding the difference between the mean of the first half vs second half
#count how many times it is larger than the mean observed before
#check if the significance against t test

m=0
for i in range(10000):
    shuffled=weeks_songlengthsorted.iloc[np.random.permutation(len(weeks_songlengthsorted))].reset_index(drop=True)
    first_group= shuffled.iloc[:(len(weeks_songlengthsorted['weekscount']))/2+1]
    first_mean=np.mean(first_group['weekscount'])
    second_group= shuffled.iloc[(len(weeks_songlengthsorted['weekscount']))/2+1:]
    second_mean=np.mean(second_group['weekscount'])
    diff_means=abs(first_mean-second_mean)
    if diff_means > .33:
        m+=1
P= m/10000.0
print m
print P
```

```python
## p value returned from shuffle test is very close to the ttest p value!! SUCCESS
```

```python
#graph seconds to weeks on charts

plt.figure(figsize=(15,10))
plt.scatter(short_list['time'],short_list['weekscount'], s=[5+x**2 for x in short_list['weekscount']], color='mediumseagreen')

plt.scatter( long_list['time'], long_list['weekscount'], s=[5+x**2 for x in long_list['weekscount']], color="mediumorchid")
plt.axvline( x=236, linewidth=2, color='dodgerblue')
plt.title('Distribution of Song Length to Weeks on Billboard', fontsize=20, y=1.09)
plt.ylabel('Weeks on Billboard', fontsize=16)
plt.xlabel('Song Length(in seconds)', fontsize=16)
plt.ylim([0,65])
plt.show()



```
